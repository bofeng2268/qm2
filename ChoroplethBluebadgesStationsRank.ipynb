```python

import pandas as pd
import plotly
import plotly.express as px
import matplotlib.pyplot as plt
import geopandas as gpd
import requests
import json
```

```python
# Paths to the files with Raw GitHub URLs
topojson_file_path = 'https://raw.githubusercontent.com/bofeng2268/qm2/main/Phenomena/Local%20Authority%20Disability/topo_lad.json'
local_authorities_file_path = 'https://raw.githubusercontent.com/Maxbruel/QM2/main/notebooks/Local%20Authorities%20London.csv'
csv_file_path = 'https://raw.githubusercontent.com/bofeng2268/qm2/main/Recent%20Acces%20Stations%20and%20Disabled%20People%202.csv'
# Load the CSV files
stations_disabled_people = pd.read_csv(csv_file_path, delimiter=';')
local_authorities = pd.read_csv(local_authorities_file_path)
#Read the CSV files into a dataframe
local_authorities_df = pd.read_csv(local_authorities_file_path, delimiter=';')
stations_disabled_people_df = pd.read_csv(csv_file_path)
# Load the TopoJSON file from the URL
response = requests.get(topojson_file_path)
topojson_data = response.json()
gdf = gpd.read_file(json.dumps(topojson_data), layer='lad')
```

```python
local_authorities.columns = local_authorities.columns.str.strip()
stations_disabled_people.columns = stations_disabled_people.columns.str.strip()
```

```python
merged_data = pd.merge(local_authorities, stations_disabled_people, left_on='Feature Name', right_on='Local Authority')

# Merge the GeoDataFrame with the merged data
final_merged_data = gdf.merge(merged_data, left_on='LAD13NM', right_on='Feature Name')

final_merged_data
```

```python
final_merged_data['Combined Metric'] = (final_merged_data['Share of transport reliant (/100)'] +
                                        final_merged_data['Normalized accesibility score (/100)']) / 2

```

```python
high_threshold = final_merged_data['Combined Metric'].quantile(0.75) # Adjust the quantile as needed
low_threshold = final_merged_data['Combined Metric'].quantile(0.25) # Adjust the quantile as needed

# Function to categorize each area
def categorize(value, high, low):
    if value >= high:
        return 'High'
    elif value <= low:
        return 'Low'
    else:
        return 'Medium'

# Apply the categorization
final_merged_data['Category'] = final_merged_data['Combined Metric'].apply(categorize, args=(high_threshold, low_threshold))
```

```python
#shows bivariate given same weight and averaged as one and their ranking (most interesting) light = low amount of people relying on transport + low borough score. dark = high accessibility score and high amount of people relying on transport here
fig = px.choropleth_mapbox(final_merged_data,
                           geojson=final_merged_data.geometry,
                           locations=final_merged_data.index,
                           color='Category',  # Assuming you have this column correctly set up in final_merged_data
                           color_discrete_map={'High': 'darkblue', 'Medium': 'blue', 'Low': 'lightblue'},
                           mapbox_style="carto-positron",
                           zoom=5,
                           center={"lat": 52.3555, "lon": -1.1743},
                           opacity=1)

fig.update_layout(margin={"r":0, "t":0, "l":0, "b":0})
fig.show()
```
